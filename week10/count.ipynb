```python
from collections import Counter, defaultdict
import numpy as np
import re

tgts = open('./pat.txt', 'r', encoding='utf8').read().split('\n\n')

counts = defaultdict(Counter)
sents = defaultdict(lambda: defaultdict(lambda: []))

for tgt in tgts:
    try:
        line, patterns = tgt.strip().split('\n', maxsplit=1)
    except:
        continue
    
    patterns = patterns.split('\n')
    
    for pattern in patterns:
        try:
            head, ptn, ngram, ch_ptn, ch_ngram = pattern.split('\t')
        except:
            continue
        
        if ptn in ['V about n', 'V in n', 'V on n', 'V to n', 'V for n', 'V with n']:
            ptn = 'V p n'
        
        # 直接 hardcode
        ch_ptn = ch_ptn.replace('n n n', 'n').replace('n n', 'n')
        
        counts[ptn][ch_ptn] += 1
        sents[ptn][ch_ptn].append("%s\t%s"%(ngram, ch_ngram))

        
def getHighCounts(key, ngrams):
    new_counts = list(map(lambda pair: (pair[0], pair[1] * (len(pair[0].split(' '))**1.5)), ngrams.items()))

    values = list(map(lambda x: x[1], new_counts))
    total, avg, std = np.sum(values), np.mean(values), np.std(values)
    
    remains = filter(lambda pair: pair[1] > avg+2*std, new_counts)
    return remains


for ngram, ch_ngrams in counts.items():
    remains = getHighCounts(ngram, ch_ngrams)

    total = np.sum(list(counts[ngram].values()))
    print("%s (%d)" % (ngram, total))
    for ch_ngram, _ in remains:
        if ch_ngram.islower():
            continue

        print("-> %s\t(%s)\t"%(ch_ngram, str(counts[ngram][ch_ngram])))
        for sent in sents[ngram][ch_ngram][:2]:
            print(sent)
    print()


```
